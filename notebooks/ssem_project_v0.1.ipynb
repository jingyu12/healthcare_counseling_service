```python
from eunjeon import Mecab

mc=Mecab()

def get_nouns(doc):
    noun_list=mc.nouns(doc)
    noun_list=[noun for noun in noun_list if len(noun)>1]
    return ' '.join(noun_list)

# 정규표현식을 사용한 노이즈 제거
def delete_noise(text):
    noise = re.compile('[\t\n\r\xa0]')                
    result=noise.sub(' ',str(text))                   #  \t, \n, \r, \xa0 제거 
    result=re.sub('[^ ㄱ-ㅣ가-힣]+',' ',result)
    result=re.sub(' +<.*?>',' ',result)               # 특수문자 제거
    result=re.sub(r'[^\w]',' ',result)                # 특수문자 제거
    result=re.sub(' +',' ',result).strip()            # 여러 공백(multi space)을 하나의 공백으로 줄이기
    return result

def preprocessing_all(doc):
    result=delete_noise(doc)
    result=get_nouns(result)
    return result
```

```python
def merge_q_title(df):
    # title을 question 앞쪽에 붙임
    df.title=df.title.apply(preprocessing_all)
    df.question=df.title+' '+df.question

    # TF Matrix를 생성하기 위해 모든 데이터 merge 함
    merged=pd.concat([df.question,survey.text]).reset_index(drop=True)
    return merged

def bigram_corpus(merged):
    # make bigram corpus
    merged=merged.reset_index(drop=True)
    splited=merged.apply(lambda x: x.split(' '))
    
    bigram = Phrases(splited, min_count=2, threshold=5)
    bigram_phraser = Phraser(bigram)
    merged2=[bigram_phraser[sent] for sent in splited]
    
    merged2=pd.Series(merged2).apply(lambda x: ' '.join(x))
    
    return merged2

merged=merge_q_title(df)
#merged=bigram_corpus(merged)
```

# Topic modeling - LDA

#### make tf

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# LDA를 위한 TF 모델 생성
tf_vec=CountVectorizer(max_df=0.95, min_df=1)
tf_model=tf_vec.fit(merged)

def make_tf(documnet_list,tf_model):
    tf_feature_names=tf_vec.get_feature_names()
    tf=tf_model.transform(documnet_list)
    print('Term frequency matrix shape :',tf.shape)
    return tf,tf_feature_names

def make_one_hot_vector(a):
    result=np.where(a==np.max(a),1,0)
    return result
```

```python
# 지식인 Question으로 TF 모델 생성a
tf_matrix,tf_feature_names=make_tf(df.question,tf_model)
# survey question으로 TF 모델 생성
s_tf_matrix,s_tf_feature_names=make_tf(survey.text,tf_model)

# merged TF
full_matrix,full_tf_feature_names=make_tf(merged,tf_model)
```

#### lda perplexity calculate

```python
def calculate_perplexity_lda(tf_for_fit,tf_for_perplx,max_topic):    
    qq=[]
    for i in range(3,max_topic):
        n_topics = i
        lda2 = LatentDirichletAllocation(n_components=n_topics,max_iter=10,
                                         learning_method='online',random_state=0, n_jobs=4).fit(tf_for_fit)
        qq.append(lda2.perplexity(tf_for_perplx))
    return qq
```

```python
qq=calculate_perplexity_lda(full_matrix,full_matrix,11)
```

```python
import itertools, operator
from gensim.models.word2vec import Word2Vec
```

```python
question_list=df.question.apply(lambda x : x.split(' '))
survey_list=survey.text.apply(lambda x : x.split(' '))

all_list=pd.concat([question_list,survey_list])
```

```python
%%time

# word2vec modeling
print('Training model...')

w_model=Word2Vec(all_list,sg=1,workers=4, min_count=1, size=300, window=10, seed = 42, iter=5, hs=1,negative=3)
#model=Word2Vec(w_corpus,sg=1,workers=4, min_count=5, size=300, window=10, seed = 42, iter=5, hs=1,negative=3)
```

```python
# using every word in test set except words not in corpus
# get embedding vectors for every words
# sum of all word vectors in one document --> document vector
# return document vector

def get_documnet_vector(question):
    doc_vec=[]
    q_len=len(question)
    for word in question:
        try:
            doc_vec.append(w_model.wv[word])  
        except:
            pass # corpus에 없는 단어는 pass
    return np.average(doc_vec,0)

# input code and get code vector 
def get_code_vector(test):
    unique_code=list(test.code.unique())
    code_vec_list=[w_model.wv[i] for i in unique_code]
    return code_vec_list
```

```python
# get infer vector 
q_vec=question_list.apply(get_documnet_vector)
s_vec=survey_list.apply(get_documnet_vector)

# make array
q_vec=np.array([i for i in q_vec])
s_vec=np.array([i for i in s_vec])
```

```python
def make_word_cloud(n_count):
    from wordcloud import WordCloud
    # font path 설정
    path = "c:/Windows/Fonts/malgun.ttf"
    wc=WordCloud(font_path=path, width=800, height=800).generate_from_frequencies(n_count)

    plt.figure(figsize=(8, 8))
    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")

def make_source_of_wordcloud(n):
    from collections import Counter
    # make flat list
    n_l=[j for i in n for j in i]
    # coount
    n_count=Counter(n_l)
    print('The number of unique nouns in question : ',len(n_count))
    return n_count
```
